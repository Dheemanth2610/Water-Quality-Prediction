```r
X <- data[, !(names(data) %in% "Potability")]
y <- data$Potability
```

```r
dim(X)
length(y)
```

```r
head(X)
```

```r
set.seed(42)

trainIndex <- sample(1:nrow(X), size = 0.5*nrow(X))

X_train <- X[trainIndex, ]
y_train <- y[trainIndex]

X_test <- X[-trainIndex, ]
y_test <- y[-trainIndex]
```

```r
library(scales)

X_train <- as.data.frame(lapply(X_train, scale))
head(X_train)
```

```r
X_test <- as.data.frame(lapply(X_test, scale))
head(X_test)
```

```r
X_train_matrix <- model.matrix(~., data = X_train)[,-1]
```

```r
y_train <- as.numeric(as.character(y_train))

lg <- glm(y_train ~ X_train_matrix)

print(summary(lg))
```

```r
pred_lg <- predict(lg, newdata = X_test)

pred_lg <- ifelse(pred_lg > 0.5, 1, 0)

res <- mean(pred_lg == y_test)
print(paste('Accuracy =', res))
```

```r
library(caret)
pred_lg<-as.factor(pred_lg)
y_test<-as.factor(y_test)
# Compute the confusion matrix
cm <- confusionMatrix(data=pred_lg,reference=y_test)

# Print the confusion matrix
print(cm)
```

```r
# Convert the confusion matrix to a table
ctable <- as.table(cm$table)

# Plot the confusion matrix
fourfoldplot(ctable, color = c("#CC6666", "#99CC99"), conf.level = 0, margin = 1, main = "Confusion Matrix")
```
